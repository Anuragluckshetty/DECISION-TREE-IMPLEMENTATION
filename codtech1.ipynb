```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
iris=load_iris()
x=iris.data
y=iris.target
```

```python
print(iris.feature_names)
print(iris.target_names)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=DecisionTreeClassifier(random_state=42)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```

```python
plt.figure(figsize=(15,8))
plot_tree(
    model,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled=True,
)
plt.show()
```

## Analysis
A Decision Tree classifier was implemented using the Iris dataset.
The model was trained on 80% of the data and tested on 20%.
The visualization clearly shows how the model splits data based on
features such as petal length and petal width.
The achieved accuracy indicates that the model performs well for
classification tasks.
